```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options 
import selenium.webdriver.support.ui as ui
import time
chrome_options = Options()  
chrome_options.add_argument("--headless")
br = webdriver.Chrome(chrome_options=chrome_options, executable_path=r'./chromedriver.exe')
```

```python
def scraping_income(city, row):
  
    br.get("https://factfinder.census.gov/faces/nav/jsf/pages/community_facts.xhtml")
    wait = ui.WebDriverWait(br,15)
    time.sleep(1.5)
    # Types search terms for a course
    element = wait.until(lambda br:br.find_element_by_id('cfsearchtextbox'))
    element.clear()
    element.send_keys(city)

    # Submits form
    element =  wait.until(lambda br: br.find_element_by_xpath('//*[@id="communityfactssubmit"]'))
    element.click()

    # Clicking the income tab
    time.sleep(1.5)
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="leftnav"]/a[7]'))
    element.click()

    # Getting table link
    time.sleep(1.5)
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="cf-content"]/div[2]/div[1]/ul[1]/li[1]/div/a'))
    href = element.get_attribute('href')
    br.get(href)
    
    #time.sleep(1)

    #for i in range(0, 5):
   #     element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="data"]/tbody/tr[79]/td[1]'))
    #    df.iloc[row, i] = element.text
   #     print('//*[@id="year_selector_content"]/ul/li['+ str(i+2) +']/a')
   #     element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="year_selector_content"]/ul/li['+ str(i+2) +']/a'))
   #     element.click()
   #     time.sleep(1.25)

   # element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="data"]/tbody/tr[79]/td[1]'))
   # df.iloc[row, 6] = element.text


    # Getting the average 2016 income
    time.sleep(1.5)
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="data"]/tbody/tr[79]/td[1]'))
    df.iloc[row, 0] = element.text

    # Clicking on 2015
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="year_selector_content"]/ul/li[2]/a'))
    element.click()

    # Getting the average 2015 income
    time.sleep(1.5)
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="data"]/tbody/tr[79]/td[1]'))
    df.iloc[row, 1] = element.text
    
    # Clicking on 2014
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="year_selector_content"]/ul/li[3]/a'))
    element.click()
    
    # Getting the average 2014 income
    time.sleep(1.5)
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="data"]/tbody/tr[79]/td[1]'))
    df.iloc[row, 2] = element.text
    
    # Clicking on 2013
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="year_selector_content"]/ul/li[4]/a'))
    element.click()
    
    # Getting the average 2013 income
    time.sleep(1.5)
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="data"]/tbody/tr[79]/td[1]'))
    df.iloc[row, 3] = element.text
    
    # Clicking on 2012
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="year_selector_content"]/ul/li[5]/a'))
    element.click()
    
    # Getting the average 2012 income
    time.sleep(1.5)
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="data"]/tbody/tr[79]/td[1]'))
    df.iloc[row, 4] = element.text
    
    # Clicking on 2011
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="year_selector_content"]/ul/li[6]/a'))
    element.click()
    
    # Getting the average 2011 income
    time.sleep(1.5)
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="data"]/tbody/tr[79]/td[1]'))
    df.iloc[row, 5] = element.text
    
    # Clicking on 2010
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="year_selector_content"]/ul/li[7]/a'))
    element.click()
    
    time.sleep(1.5)
    element = wait.until(lambda br:br.find_element_by_xpath('//*[@id="data"]/tbody/tr[79]/td[1]'))
    df.iloc[row, 6] = element.text
```

# The next cell is where we get the list of the cities that we are working with.


```python
cities = [line.rstrip('\n\r') for line in open('./cities.txt')]

df = pd.DataFrame(columns=["City", "Mean Income 2016", "Mean Income 2015", "Mean Income 2014", "Mean Income 2013",
                  "Mean Income 2012", "Mean Income 2011", "Mean Income 2010"])
data = pd.DataFrame({"City": range(len(cities))})
df = df.append(data)

for i, row in df.iterrows():
    row.loc['City'] = cities[i]

# Reverse the column order (ie 2010 to 2016)
df = df.iloc[:, ::-1]

for i, city in enumerate(cities):
    scraping_income(city, i)
    print(city+str(i))
    display(df)
    

df = df.set_index('City')

df.to_csv('cities_income.csv', encoding='utf-8', index=False)
br.quit()
```
